```python
# Plotting multiple curves 

x = np.arange(-10,20)
y = x **2

x2 = np.arange(0,30)
y2 = x2 + 50

plt.plot(x,y,'g--')
plt.plot(x2,y2)

plt.text(7.5,60,'Hello')

plt.show()
```

```python
plt.plot(x,y,'',x2,y2,'') # we can plot mutiple plots in a single line
```

```python
# Plotting multiple curves 

x = np.arange(-10,20)
y = x **2

x2 = np.arange(0,30)
y2 = x2 + 50

plt.plot(x,y,'g--',label='quadratic')
plt.plot(x2,y2,label='linear')

plt.text(7.5,60,'Hello')

plt.legend() # after this only, label will be displayed
plt.show()
```

## Logarithmic graph

```python
x = np.arange(1,100)
y = np.log(x) # but its by chance available, so we can take from math library

plt.plot(x,y)
plt.show()
```

```python
import math

x = np.arange(1,100)
#y = math.log(x) This is wrong as we have pass list into it , where as we have to pass a number

y= [math.log(i) for i in x]

plt.plot(x,y)
plt.show()
```

```python
import math

x = np.arange(1,100)
y= list(map(lambda i : math.log(i),x))

plt.plot(x,y)
plt.show()
```

## Vectorize function in numpy

```python
import math
x = np.arange(1,100)
#foo = np.vectorize(math.log)
#y = foo(x)
# OR

y = np.vectorize(math.log)(x)

plt.plot(x,y)
plt.show()
```

## Histogram

```python
# here we will raw data and it will itself figure out and also the height of bar.

vals = [1,2,3,2,3,2,1,3,2,4,2,3,2]
plt.hist(vals)
plt.show() # here it has automatically calculated it's frquency for Y-axis
```

```python
vals = [1,2,3,2,3,2,1,3,2,4,2,3,2]
plt.hist(vals,[0,2,3,4,10])
plt.show() # we have defines our own buckets -- 0-2, 2-3,3-4,4-10 (starting point included)
```

```python
vals = [1,2,3,2,3,2,1,3,2,4,2,3,2]
plt.hist(vals,3) # Here 3 means that it should divide  it in 3 buckets and it will divide it by itself
plt.show()
```

## Sigmoid

```python
x = np.arange(-10,11)

sigmoid = lambda  x : 1/(1+math.exp(-x))

y = [sigmoid(i) for i in x]

plt.plot(x,y)
plt.show()
```
